```python
def sum_5_df(arr):
    mean = np.mean(arr)
    std = np.std(arr)
    q1 = np.quantile(arr, 0.25)
    q2 = np.median(arr)
    q3 = np.quantile(arr, 0.75)
    minimum = np.min(arr)
    maximum = np.max(arr)
    total = arr.sum()
    table = {
        'metric': [arr.name],
        'mean': [mean],
        'std': [std],
        'min': [minimum],
        'q1': [q1],
        'median': [q2],
        'q3': [q3],
        'max': [maximum],
        'total': [total]
    }
    table = pd.DataFrame(table)
    return table
```

```python
rebounds_list = [sum_5_df(rebounds_df[col]) for col in rebounds_df.columns]
rebounds_sum = pd.concat(rebounds_list, ignore_index=True)
```

```python
model_formula_1 = 'result_W ~ Q("3P%_gsw") + Q("2P%_gsw") + Q("FT%_gsw") + ORB_gsw + DRB_gsw + AST_gsw + \
                         STL_gsw + BLK_gsw + TOV_gsw + PF_gsw + location_home'
```

```python
import statsmodels.formula.api as smf
logit_model = smf.logit(model_formula_1, data=train_set).fit()
```

```python
simulated_data
```

```python
def predict_model(model, x_test, y_test):
    y_pred_probs = model.predict(x_test)
    y_pred_classes = (y_pred_probs >= 0.5).astype(int)
    cm = confusion_matrix(y_test, y_pred_classes)
    print(cm)
    sns.heatmap(cm/np.sum(cm), annot=True, fmt='.2%', cmap='Blues', xticklabels=['Loss', 'Win'], yticklabels=['Loss', 'Win'])
    print(classification_report(y_test, y_pred_classes))
```

```python
predict_model(logit_model_2, simulated_data.iloc[:, 0:13], simulated_data.iloc[:, -1])
```
