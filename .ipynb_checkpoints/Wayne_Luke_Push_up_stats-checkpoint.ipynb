```python
import re
def clean_value(val):
    match = re.match(r'^(\d+(\.\d+)?)', val)
    if match:
        num = match.group(1)
        return float(num) if '.' in num else int(num)
    else:
        return None
```

```python
df['cleaned_value'] = df['value'].apply(clean_value)
```

```python
df.cleaned_value.unique()
```

```python
df.day.unique()
```

```python
def clean_day(val): 
    val = re.sub(r'\s+', '', val) # Replace multiple spaces with a single space (required as it will not match properly) 
    match = re.search(r'(\d+)$', val)
    if match:
        return int(match.group(1))
    else:
        return None
```

```python
df['cleaned_day'] = df['day'].apply(clean_day)
```
