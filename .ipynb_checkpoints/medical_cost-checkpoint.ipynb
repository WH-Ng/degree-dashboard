### Two-sample Hypothesis Test (Independent Sample T-test) - Smoker Status vs Charges

Let $\mu_A$ = The mean amount of insurance claim for smokers.

Let $\mu_B$ = The mean charges/amount of insurance claim for non-smokers

$$H_0: \mu_A = \mu_B$$

$$H_1: \mu_A \neq \mu_B$$

In words:
- Null Hypothesis states: There is no difference in amount of insurance claim between smokers and non-smokers.
- Alternative Hypothesis states: There is a difference in amount of insurance claim between smokers and non-smokers

Conducting this test at $\alpha$ = 0.05

```python
smoker_df = df[df.smoker == 'yes']
non_smoker_df = df[df.smoker == 'no']
smoker_charges = df[df.smoker == 'yes'].charges
non_smoker_charges = df[df.smoker == 'no'].charges
```

```python
# Checking whether to use equal variance
round(np.std(smoker_charges) / np.std(non_smoker_charges), 2)

# Since ratio is less than 2, using equal variance here is fine. Python defaults to equal variance
```

```python
t_out = stats.ttest_ind(smoker_charges, non_smoker_charges)
```

```python
ci_95 = t_out.confidence_interval()
```

```python
print(f"""
* The p-value for the test is {t_out.pvalue:.3f}. 
* The value of the test statistic is {t_out.statistic:.3f}.
* The upper and lower limits of the CI are ({ci_95[0]:.0f}, {ci_95[1]:.0f}).
""")
```

#### Conclusion

Since the p-value (0.00) is less than 0.05, we have strong evidence to reject the null hypothesis. We are 95% confident that there is mean difference between the charges of insurance claim between smokers and non-smokers

### Wilcoxon Rank Sum (WRS) Test - Smoker Status vs Charges

```python
wrs_out = stats.mannwhitneyu(smoker_charges, non_smoker_charges)

print(f"""Test statistic: {wrs_out.statistic:.0f}.
p-val: {wrs_out.pvalue:.3f}.""")
```

#### Assumptions
- The sample sizes of both smoker and non-smokers are at least 10, and come from an underlying continuous distribution
- Hence, assumptions hold and the test statistic follows a N(0, 1) distribution

#### Conclusion

Since the p-value (0.000) is less than 0.05, we have strong evidence to reject the null hypothesis. Hence, there is a significance  difference between the mean charges of insurance claim between smokers and non-smokers

### Kruskal-Wallis Procedure - Region vs Charges

```python
df.region.unique()
```

Let $\mu_i$ = The mean of charges for each unique region. For $i = 1, 2, 3, 4$

$$H_0: \mu_1 = \mu_2 = \mu_3 = \mu_4$$ 

$$H_1: \text{At least one of the $\mu_i$ is different from the others}$$

In words:
- Null Hypothesis states: There is no difference in mean amount of insurance claim between all the unique region
- Alternative Hypothesis states: There is at least one unique region that is different from the other region in the mean amount of insurance claim

Conducting this test at $\alpha$ = 0.05

Additonally, we can further ask if there is any difference between the mean amount of northeast and northeast. Finally, we want to know if they is any difference between the mean amount of group A (southwest and southeast) and group B (northwest and northeast)

```python
df.groupby('region')['charges'].count()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
out = [x[1] for x in df.charges.groupby(df.region)]
```

```python
kw_out = stats.kruskal(*out)
```

```python
print(f"""The test statistic is {kw_out.statistic:.3f}, the p-value is {kw_out.pvalue:.3f}.""")
```

#### Conclusion

Since the p-value of 0.192 is greater than 0.05, we have insufficient evidence to reject the null hypothesis. Hence, we conclude that themean amount of charges in each region is the same.
